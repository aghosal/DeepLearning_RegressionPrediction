The Data showed number of columns has more than 40% Missing value, which may compromise the the model if used after imputation. Thus I should remove any column that has more than 40% NAs.

```python
#Find the columns that have >40% Null values
NAN_threshold = 0.4
col_name = []
Missing_percent = []

for col in df.columns:
    NA_count = (df[col].isna().sum() / df.shape[0])*100
    if NA_count > NAN_threshold:
        col_name.append(col)
        Missing_percent.append(NA_count)

NA_col = pd.DataFrame({'col': col_name, "%-Missing": Missing_percent})

print(f'The total numbers of Columns with >50% NAs is {len(NA_col)}')
NA_col.head(50).sort_values(by='%-Missing',ascending=False).head(10)
```

```python
#Drop Colume where NA>40%
df1=df.drop(columns=col_name,axis=1)#REmoving the features that showed >40% NAs

df1_table=pd.DataFrame(df1.isna().sum().sort_values(ascending=False)).reset_index()
df1_table.columns=['Feature_Name','Missing Value']
df1_table['%']=(df1_table['Missing Value']/df1.shape[0])*100

print(f"The number of remaining features are:{len(df1_table)}")
df1_table.head(n=20)
```

```python
# Split x and y
from sklearn.model_selection import train_test_split
x=df1.drop(columns=['TARGET'],axis=1)
y=df1['TARGET']
```

```python
# Here I filledin the columns with corresponding mean for the numerical and 'most_frequest' for categorical
from sklearn.impute import SimpleImputer

x_num_col=x.select_dtypes(exclude='object').columns.to_list()
x_cat_col=x.select_dtypes(include='object').columns.to_list()

SI = SimpleImputer(strategy='mean')
x[x_num_col] = pd.DataFrame(SI.fit_transform(x[x_num_col]), columns=x_num_col,index=x.index)

SC = SimpleImputer(strategy='most_frequent')
x[x_cat_col] = pd.DataFrame(SC.fit_transform(x[x_cat_col]), columns=x_cat_col,index=x.index)
print(f"Total number of NAs is: {x.isna().any().sum()}")# To confirm that all missing values are replaced with corresponding imputation value.
x.head(5)

```

```python
# Winsorize the outliers before Minmax scaling, since there are too many column I feel safe to normalize by winsorize instead of looking individual columns
from scipy.stats.mstats import winsorize

for col in x_num_col:
  x[col]=winsorize(x[col],limits=[0.05,0.05])# All data within 95% of the discrtibution will be included
```

```python
from sklearn.preprocessing import MinMaxScaler

# Fit scaler on training numeric columns
mm = MinMaxScaler()#Since I removed the outlier the chances of skewing the scale is less, I used MinMax instead of standard normalization as this goes well with Deep learning
x[x_num_col] = pd.DataFrame(mm.fit_transform(x[x_num_col]),columns=x_num_col)
```

```python
# cardinality of each columns for Categorical columns for encoding
for col in x[x_cat_col]:
  print(f"Number of unique values for {col} in {x[col].nunique()}")
  print(f"Number of unique values for {col} in {x[col].unique()}")
```

ORGANIZATION TYPE has 58 unique values which is not ideal for one-hot encoding, thus I used label encoding for 'ORGANIZATION TYPE' while performed one-hot-encoding for for rest of the categorical variables.

```python
from sklearn.preprocessing import LabelEncoder# Label encoding for ORGANIZATION TYPE
le=LabelEncoder()
x['ORGANIZATION_TYPE']=le.fit_transform(x['ORGANIZATION_TYPE'])
```

```python
x['ORGANIZATION_TYPE']# The Label Encoding is sucessfully done, since the colume is already numeric it will not be affeceted during one hotencoding
```

```python
cat_cols = x.select_dtypes(include='object').columns.tolist()# One Hot encoding for remaining categorical variables.

x = pd.get_dummies(x, columns=cat_cols, dtype=int)
```

```python
x.columns# One-hot-encoding is done scuccessfully
```

```python
# In order to leveling up the minority class I split the data first and mane sure I over sample to train dataset, and test data remains untouched.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)

from imblearn.over_sampling import ADASYN
adasyn=ADASYN(random_state=42, sampling_strategy='minority',n_neighbors=5)
x_train_resample,y_train_resample=adasyn.fit_resample(x_train,y_train)
```

```python
import matplotlib.pyplot as plt# The over samplified samples between imbalanced and balanced set
fig,axes=plt.subplots(1,2,figsize=(8,6))
y_test.value_counts().plot(kind='bar',ax=axes[0],title='Initial Sample Distribution')
y_train_resample.value_counts().plot(kind='bar',ax=axes[1],title='Sample distribution after ADASYN')
plt.tight_layout()
```


Here we see the minority class is increased to the same level as majority class. Observation: When 5 neighbors were used there were more chances to find the "hard region" near the boundaries, while for n_neighbors=2, since there are n_neighbors, we get less synthetically generated minority class. Thus, I noticed less number of minority class for n_neighbor=2. After fitting I noticed n_neighbor=5 gives the best fittings so in the final model I am using n=5.

## 7) Calculate Sensitivity as a metric; 8) Calculate the area under the ceceiver operative characterics curve

```python
import tensorflow as tf
from tensorflow.keras import models,layers,metrics
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import roc_auc_score,classification_report,recall_score
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
from tensorflow.keras.callbacks import ModelCheckpoint

# Setup Architecture
nn_model=models.Sequential([layers.Input(shape=(x_train_resample.shape[1],)),
                            layers.GaussianNoise(stddev=0.1),#Introducing noise for generalization, similar to data augmentation in CNN

                            layers.Dense(128,kernel_regularizer=l2(0.0001)),
                            layers.BatchNormalization(),#Provides stabilization of the model
                            layers.LeakyReLU(negative_slope=0.01),#seems like better than Relu for this dataset, mstly because it is highly imbalanced.
                            layers.Dropout(0.1),

                            layers.Dense(64,kernel_regularizer=l2(0.0001)),
                            layers.BatchNormalization(),
                            layers.LeakyReLU(alpha=0.01),
                            layers.Dropout(0.1),

                            layers.Dense(32),
                            layers.BatchNormalization(),
                            layers.LeakyReLU(alpha=0.01),
                            layers.Dropout(0.1),

                            layers.Dense(1,activation='sigmoid')])

checkpoint = ModelCheckpoint('best_nn_model.keras',
                             monitor='val_loss',
                             save_best_only=True,
                             mode='min',
                             verbose=1)

reduce_lr = ReduceLROnPlateau(monitor='val_loss',factor=0.5, patience=3,min_lr=1e-6,verbose=1)

early_stop = EarlyStopping(monitor='val_loss',
                           patience=3,
                           restore_best_weights=True)

#compile model
nn_model.compile(optimizer=Adam(learning_rate=0.0005),
                 loss='binary_crossentropy',
                 metrics=[tf.keras.metrics.Recall(name='sensitivity'),tf.keras.metrics.AUC(name='roc_auc'),tf.keras.metrics.AUC(name='pr_auc', curve='PR')])# Precision-Recall curve
nn_model.summary()


#Train the model

history=nn_model.fit(x_train_resample,y_train_resample,
                     validation_split=0.2,
                     epochs=50,
                     batch_size=32,#TEsted 64 as well, looks like 32 is little better
                     callbacks=[early_stop, reduce_lr,checkpoint],#TO make sure the model stop fitting before model stops learning and starts over fitting
                     verbose=1,
                     shuffle=True)
```

```python
# Test the model on the test dataset
from sklearn.metrics import roc_curve
y_pred_prob=nn_model(x_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
j_scores = tpr - fpr
best_threshold = thresholds[j_scores.argmax()]
roc_auc = roc_auc_score(y_test, y_pred_prob)
y_pred_prob=nn_model.predict(x_test)
y_pred = (y_pred_prob >= best_threshold).astype(int)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"Sklearn ROC-AUC Score: {roc_auc:.4f}")
print(f"Optimal Threshold by Youden’s J: {best_threshold:.4f}")

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Recall)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The mode showed the AUC at 0.66 and the recall for negative samples are 0.58 and for postive group it showed 0.66. THis means the data is sensitive to the level of 0.66, however the specificity or precision is low. That means while doing the threshold it mis catagorised some samples as False Positive.

In order to evaluate I also tried to do with .evaluate but did not get good performance on the model. In this method 0.5 is used by default so I tried to find Youdon's statistics which ultimately gave the best perfromance. In imbalanced data normally a threshold of 0.2-0.3 is the expected.In fact I got the best threshold from Youdon' J statistics in this range.

Below I am trying to look for the baseline in the standard MAchine learning to get idea if the model can be made any better.

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,roc_auc_score, recall_score, roc_curve
import matplotlib.pyplot as plt
import numpy as np

# Define models
models = {"RandomForest": RandomForestClassifier(n_estimators=500, random_state=42),"GradientBoost": GradientBoostingClassifier(random_state=42),"LogisticRegression": LogisticRegression(max_iter=500, random_state=42)}
# Plot settings
plt.figure(figsize=(8, 6))

# Loop through models
for name, model in models.items():
    print(f"\nTraining and evaluating: {name}")
    
    # Train
    model.fit(x_train_resample, y_train_resample)
    
    # Predict probabilities
    y_pred_proba = model.predict_proba(x_test)[:, 1]
    
    # ROC Curve
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    
    # Youden's J statistic
    youden_j = tpr - fpr
    optimal_idx = np.argmax(youden_j)
    optimal_threshold = thresholds[optimal_idx]
    print(f"Threshold by Youden's J for {name}: {optimal_threshold:.4f}")
    
    # Predict using optimal threshold
    y_pred = (y_pred_proba >= optimal_threshold).astype(int)
    
    # Sensitivity (Recall)
    sensitivity = recall_score(y_test, y_pred)
    print(f"Sensitivity (Recall) for {name}: {sensitivity:.4f}")
    
    # ROC AUC
    auc_score = roc_auc_score(y_test, y_pred_proba)
    print(f"ROC AUC Score for {name}: {auc_score:.4f}")

```

From the above model it shows that Random forest could do similar like DL, so in the code below I used hyper parameter tuning on the Randome Forest and compare the model performances. 
The similar performances between ML and DL, I think happen becaue of the following based on my background research, 1) Since the data is highly imbalanced there may be need of more positive cases, 2)There might be enough linear relation present to seperate the classes (so DL could not shine), 3) although I used trial-error using different hyper parameter in tuning we may not have reached super tuned model in DL yet. The difference however between th DL and RandoForst is not too high to suggest that there might be not enough scope for tuning in DL further.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV

# 1. Define parameter grid for Randomized Search
param_grid = {'n_estimators': [100, 300, 500],'max_depth': [None, 10, 20, 30],'min_samples_split': [2, 5, 10],'min_samples_leaf': [1, 2, 4],'max_features': ['sqrt', 'log2'],'bootstrap': [True, False]}

# 2. Setup StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rf = RandomForestClassifier(random_state=42)
rf_cv = RandomizedSearchCV(rf,param_grid,n_iter=20,scoring='recall',cv=skf,verbose=1,n_jobs=-1,random_state=42)

# 4. Fit on resampled training data
print("\nTuning Random Forest with StratifiedKFold CV...")
rf_cv.fit(x_train_resample, y_train_resample)

# 5. Get best model and parameters
best_rf = rf_cv.best_estimator_
print("\nBest Hyperparameters:")
print(rf_cv.best_params_)

# 6. Evaluate on test set
y_pred_proba = best_rf.predict_proba(x_test)[:, 1]

# ROC curve and Youden’s J
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
youden_j = tpr - fpr
optimal_idx = np.argmax(youden_j)
optimal_threshold = thresholds[optimal_idx]

# Predict using Youden's optimal threshold
y_pred = (y_pred_proba >= optimal_threshold).astype(int)

# Metrics
sensitivity = recall_score(y_test, y_pred)
auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"\nOptimal Threshold (Youden's J): {optimal_threshold:.4f}")
print(f"Sensitivity (Recall): {sensitivity:.4f}")
print(f"ROC AUC Score: {auc_score:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Random Forest (Tuned) Confusion Matrix")
plt.show()

# ROC Curve Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"RandomForest (AUC = {auc_score:.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Tuned Random Forest")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

The performance did not change too much, holding the moldel for Neural Network.
